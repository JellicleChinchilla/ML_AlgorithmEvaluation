```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pima.csv")
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
df_cfs = pd.read_csv("pima-CFS.csv")
df_cfs.head()
```

```python
df_cfs.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.title('Correlation Matrix without CFS Feature Selection')
plt.savefig("Correlation_Matrix.png")
plt.show()
```

```python
# compute correlation matrix
corr_matrix = df.corr()

# extract upper triangular part of the matrix (excluding diagonal)
corr_values = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# compute mean of the correlation coefficients
mean_corr = corr_values.stack().mean()

print("Average correlation coefficient: {:.2f}".format(mean_corr))
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_cfs.corr(), annot=True, cmap='Blues')
plt.title('Correlation Matrix with CFS Feature Selection')
plt.savefig("Correlation_Matrix_CFS.png")
plt.show()
```
