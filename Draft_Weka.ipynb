```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold

# Load the original Pima dataset
df = pd.read_csv("pima.csv")

# Define the number of folds
n_folds = 10

# Use StratifiedKFold to split the data into 10 folds
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Create a new DataFrame to hold the folds
fold_df = pd.DataFrame(columns=["Fold", "Contents"])

# Iterate over the folds and store them in fold_df
for fold_idx, (train_idx, test_idx) in enumerate(skf.split(df, df["class"])):
    # Get the name of the current fold
    fold_name = f"fold{fold_idx + 1}"
    
    # Get the contents of the current fold
    fold_contents = "\n".join(df.iloc[test_idx].astype(str).apply(lambda x: ",".join(x), axis=1).values)
    
    # Add the fold to fold_df
    fold_df = fold_df.append({"Fold": fold_name, "Contents": fold_contents}, ignore_index=True)

# Save fold_df to pima-folds.csv
fold_df.to_csv("pima-folds.csv", index=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def classify_nn(training_filename, testing_filename, k):
    with open(training_filename, 'r') as f:
        training_data = []
        for line in f:
            example = tuple(line.strip().split(','))
            training_data.append(example)
        X_train = []
        y_train = []
        for example in training_data:
            X_train.append(list(map(float, example[:-1])))
            y_train.append(example[-1])
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    with open(testing_filename, 'r') as f:
        testing_data = []
        for line in f:
            example = tuple(line.strip().split(','))
            testing_data.append(example)
        X_test = []
        for example in testing_data:
            X_test.append(list(map(float, example)))
    y_pred = knn.predict(X_test)
    counts = {}
    max_count = 0
    for label in y_pred:
        if label in counts:
            counts[label] += 1
        else:
            counts[label] = 1
        if counts[label] > max_count:
            max_count = counts[label]
    if len(counts) == 1:
        results = [str(label) for label in y_pred]
    else:
        max_labels = []
        for label, count in counts.items():
            if count == max_count:
                max_labels.append(label)
        if len(max_labels) == 1:
            results = [str(label) for label in y_pred]
        else:
            results = []
            for label in y_pred:
                if label in max_labels:
                    results.append(str(label))
                else:
                    results.append('Yes')
    return results
```

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

def classify_nb(training_filename, testing_filename):
    with open(training_filename) as f:
        training_data = []
        for line in f.readlines():
            training_data.append(line.strip().split(","))
        X_train = []
        y_train = []
        for row in training_data[1:]:
            X_train.append(list(map(float, row[:-1])))
            y_train.append(row[-1])
        X_train = np.array(X_train)
        y_train = np.array(y_train)
        clf = GaussianNB()
        clf.fit(X_train, y_train)
    with open(testing_filename) as f:
        testing_data = []
        for line in f.readlines():
            testing_data.append(line.strip().split(","))
        X_test = []
        for row in testing_data:
            X_test.append(list(map(float, row)))
        X_test = np.array(X_test)
        y_pred = clf.predict_proba(X_test)
        results = []
        for prediction in y_pred:
            if prediction[0] > prediction[1]:
                results.append("no")
            elif prediction[0] < prediction[1]:
                results.append("yes")
            else:
                results.append("yes")
        return results
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load the Pima Indians Diabetes dataset
pima = pd.read_csv('pima.csv')
X = pima.drop('class', axis=1).values
y = pima['class'].values

# Define the models
models = [('1-NN', KNeighborsClassifier(n_neighbors=1)),
          ('5-NN', KNeighborsClassifier(n_neighbors=5)),
          ('Naive Bayes', GaussianNB())]

# Perform 10-fold cross-validation
kfold = KFold(n_splits=10, shuffle=True, random_state=3308)
for name, model in models:
    scores = []
    for train_index, test_index in kfold.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        scores.append(accuracy_score(y_test, y_pred))
    print(f'{name} accuracy: {np.mean(scores):.3f} (std: {np.std(scores):.3f})')
```
